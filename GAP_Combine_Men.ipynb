```python
def format_price(price):
    # Check if price contains a dollar sign and is in the form '10.00$14.95'
    if '$' in price:
        # Split the price on '$', add the dollar sign to each price, and join with ' to '
        formatted_price = ' to '.join([f"${p}" for p in price.split('$')])
    else:
        # Format as a single price with a dollar sign
        formatted_price = f"${price}"
    return formatted_price

# Apply the formatting function to both columns
men_append['Selling_Price'] = men_append['Selling_Price'].apply(format_price)
men_append['Max_Retail_Price'] = men_append['Max_Retail_Price'].apply(format_price)
```

```python
def handle_price(price_range):
    # Check if the price range is a string and contains the '$' character
    if isinstance(price_range, str) and '$' in price_range:
        prices = price_range.split('$')
        # Try to convert the prices to float, if not possible due to formatting, use None
        try:
            return [float(price) for price in prices]
        except ValueError:
            return [None, None]
    else:
        # If the price range is not in the expected format, return None values
        return [None, None]
```

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re
from scipy.stats import spearmanr
import matplotlib.pyplot as plt
```

```python
# Data cleaning: handle missing values and clean text
men_append['Product_Details'] = men_append['Product_Details'].fillna('').apply(lambda x: re.sub(r'[^\w\s]', '', x.lower()))

# Display the first few rows to verify
print(men_append.head())
```

```python
# Define stopwords
stop_words = set(stopwords.words('english'))

# Function to preprocess and tokenize text
def preprocess_text(text):
    tokens = word_tokenize(text.lower())  # Convert to lower case and tokenize
    filtered_tokens = [word for word in tokens if word.isalpha() and word not in stop_words]
    return filtered_tokens

# Apply the preprocessing to each row in the 'Product_Details' column
tokens = men_append['Product_Details'].dropna().apply(preprocess_text)  # Drop NA to avoid errors

# Flatten the list of tokens and count occurrences
all_tokens = [token for sublist in tokens for token in sublist]
token_counts = Counter(all_tokens)

# Get the 10 most common tokens
most_common_tokens = token_counts.most_common(10)
# Print or return the most common tokens
print("The 10 most frequent keywords are:")
for word, count in most_common_tokens:
    print(f"{word}: {count}")
```

```python
cotton_data = men_append[men_append['Product_Details'].str.contains('cotton', case=False, na=False)]
```

```python
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import CountVectorizer
```

```python
def preprocess_text(text):
    tokens = word_tokenize(text.lower())  # Convert to lower case and tokenize
    filtered_tokens = [word for word in tokens if word.isalpha() and word not in stop_words]
    return ' '.join(filtered_tokens)  # Join the tokens back into a single string

# Apply preprocessing to each row in the 'Product_Details' column
cotton_data['processed_text'] = cotton_data['Product_Details'].apply(preprocess_text)

# Vectorize the processed text
vectorizer = CountVectorizer(max_df=0.95, min_df=2, stop_words='english')
data_vectorized = vectorizer.fit_transform(cotton_data['processed_text'])

# Apply LDA
lda_model = LatentDirichletAllocation(n_components=5, random_state=0)
lda_model.fit(data_vectorized)

# Function to display topics
def display_topics(model, feature_names, no_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print("Topic %d:" % topic_idx)
        print(" ".join([feature_names[i] for i in topic.argsort()[:-no_top_words - 1:-1]]))

# Display the topics
display_topics(lda_model, vectorizer.get_feature_names_out(), 10)
```

```python
# Expanded list of eco-friendly keywords including your specific additions
eco_keywords = [
    'recycled', 'organic', 'biodegradable', 'natural', 'sustainable',
    'fair trade certified', 'gots', 'oeko-tex', 'fsc', 'leed certified',
    'carbon-neutral', 'eco-friendly dyes', 'low impact dye', 'water-saving',
    'energy-efficient', 'reusable', 'compostable', 'upcycled', 'zero waste',
    'eco-conscious', 'environmentally friendly', 'green', 'non-toxic', 'ethical',
    'washwell', 'water', 'forever', 'always', 'responsibly', 'less', 'waste', '100'
]
```

```python
# Function to identify presence of eco-friendly keywords
def is_eco_friendly(details):
    # Tokenize and convert to lowercase to ensure case-insensitive matching
    tokens = [token.lower() for token in word_tokenize(details)]
    return 1 if any(keyword in tokens for keyword in eco_keywords) else 0

# Apply the function to create a new binary column
men_append['Eco_Friendly'] = men_append['Product_Details'].apply(is_eco_friendly)

# Check how many products are eco-friendly
print("Eco-friendly products count:", men_append['Eco_Friendly'].sum())

# Convert ratings to numeric and handle missing values
men_append['Rating'] = pd.to_numeric(men_append['Rating'], errors='coerce').fillna(0)
men_append['Rating_Count'] = men_append['Rating_Count'].replace('Not available', 0).astype(int)
```

```python
# Spearman correlation
correlation_rating, p_value_rating = spearmanr(men_append['Eco_Friendly'], men_append['Rating'])
correlation_count, p_value_count = spearmanr(men_append['Eco_Friendly'], men_append['Rating_Count'])

print(f"Spearman Correlation between Eco-Friendly and Rating: {correlation_rating} (p-value: {p_value_rating})")
print(f"Spearman Correlation between Eco-Friendly and Rating Count: {correlation_count} (p-value: {p_value_count})")
```

```python
from scipy.stats import f_oneway

# Separate ratings into two groups: eco-friendly and non-eco-friendly
ratingcounts_eco = men_append[men_append['Eco_Friendly'] == 1]['Rating_Count']
ratingcounts_non_eco = men_append[men_append['Eco_Friendly'] == 0]['Rating_Count']

# Perform one-way ANOVA
f_statistic, p_value_anova = f_oneway(ratingcounts_eco, ratingcounts_non_eco)

print(f"One-way ANOVA F-statistic: {f_statistic}")
print(f"P-value: {p_value_anova}")

# Interpret the results
if p_value_anova < 0.05:
    print("There is a statistically significant difference in rating counts between eco-friendly and non-eco-friendly products.")
else:
    print("There is no statistically significant difference in rating counts between eco-friendly and non-eco-friendly products.")
```
